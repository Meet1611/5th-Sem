### 2)  Identify Nominal, Ordinal, Binary and Numeric attributes from data sets and display all values.



```python
nomial_features = ['Name', 'Sex', 'Ticket', 'Cabin', 'Embarked']
ordinal_features = ['Pclass']
binary_features = ['Survived']
numerical_features = ['Age', 'SibSp', 'Parch', 'Fare']

print(f'Nominal Features : ',nominal_features)
print(f'Ordinal Features : ',ordinal_features)
print(f'Binary Features : ',binary_features)
print(f'Numerical Features : ',numerical_features)
```

### 4)  For each quantitative attribute, calculate its average, standard deviation, minimum, mode, range and maximum values.



```python
quantitative_features = df.select_dtypes(include=['number']).columns.tolist()
for feature in quantitative_features:
    print(feature + ':')
    print('\t Mean = ', df[feature].mean().round(2))
    print('\t Standard deviation = ', df[feature].std().round(2))
    print('\t Minimum = ', df[feature].min())
    print('\t Maximum = ', df[feature].max())    
```

### 9) Display the histogram for Age attribute by discretizing it into 8 separate bins and counting the frequency for each bin.

```python
import matplotlib.pyplot as plt
```

```python
df['Age'].plot(kind='hist', bins=8)
plt.grid()
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Age Distribution')
```
